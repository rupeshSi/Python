```python
# https://github.com/mwaskom/seaborn-data  WE can connect from githum seaborn data

tips=sns.load_dataset('tips')
```

```python
print(tips.head())
```

```python
#Distribution plot
#Univariate Distribution
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['tip'])
```

# jointplot

jointplot() allows you to basically match up two distplots for bivariate data. With your choice of what kind parameter to compare with:

    “scatter”
    “reg”
    “resid”
    “kde”
    “hex”


```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
#Regression of x on Y

sns.jointplot(x='total_bill',y='tip',data=tips,kind='resid')
```

```python
#Kernel density estimation 
#Shows maximum density of the data


sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

### pairplot

pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns).


```python
sns.pairplot(tips)
```

```python
#Pass categorical variable to hue to get the plot displayed in different colours

sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips,hue='smoker')
```

```python
sns.pairplot(tips,markers=["D", "o"],hue='sex' )
```

```python
sns.pairplot(tips,markers=["D", "o"],hue='sex', palette='twilight_r' )
```

```python
sns.pairplot(tips,markers=["D", "o"],hue='sex', palette='copper' )
```
